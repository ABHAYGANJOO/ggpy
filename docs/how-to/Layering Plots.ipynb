```python
from ggplot import *
%matplotlib inline
```

```python
ggplot(diamonds, aes(x='carat', y='price'))
```

```python
ggplot(diamonds, aes(x='carat', y='price')) + geom_point()
```

```python
ggplot(diamonds, aes(x='carat', y='price')) + geom_point() + stat_smooth()
```

```python
ggplot(diamonds, aes(x='carat', y='price')) + geom_point() + stat_smooth() + ggtitle("Diamond Carat vs. Price")
```
